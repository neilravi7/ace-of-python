```python
# Python Varibales scopes Global, Local, Nonloacal

global_variable = 0
def show():
    global_variable = 1 # same name but diffrent varibale in fuction local scope.
    print(global_variable)

show()
print(global_variable) # Value is still 0

# Uses of non local
def outer_display():
    variable_x = 1
    def inner_display():
        nonlocal variable_x
        variable_x = 2
        print("From inner display", variable_x)
    inner_display()

    print(variable_x, " from outer_display")

display()
print(global_variable)
outer_display()
print(__name__)

```

```python
# Operator overloading in python 

"""_summary_
    Function	Description
__init__()	Initialize the attributes of the object.
__str__()	Returns a string representation of the object.
__len__()	Returns the length of the object.
__add__()	Adds two objects.
__call__()	Call objects of the class like a normal function.
"""

# add two objects without operator overloading
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def add_point(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)
    
p1 = Point(3, 4)
p2 = Point(5,6)

p3 = p1.add_point(p2)
# print(p3.x, p3.y) output => 8 10

# add two objects with operator overloading
class Points:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Points(x, y)
    
p1 = Points(10, 20)
p2 = Points(30, 40)

p3 = p1+p2
# print(p3.x, p3.y) output => 40 60

# Rises of the new method
class Representation(object):
    def __new__(cls, *args, **kwargs):
        print("from new")
        print(cls)
        print(args)
        print(kwargs)

        # create our object and return it.
        obj = super().__new__(cls)
        return obj
    
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
        print("from init")

r1 = Representation(10, 20)

class SqPoint(Point):
    Max_inst = 4
    Inst_created = 0

    def __new__(cls, *args, **kwargs):
        if cls.Inst_created >= cls.Max_inst:
            raise ValueError("Can not create more objects")
        cls.Inst_created +=1    
        return super().__new__(cls)
    

p1 = SqPoint(0,0)
p2 = SqPoint(1,0)
p3 = SqPoint(1,1)
p4 = SqPoint(0,1)
# p5 = SqPoint(1,2) cannot create more objects
```

```python
# Python Decorators

# Nested functions
def outer():
    print("I am an outer function")
    def inner():
        print("I am inner function")
    return inner

the_caller = outer()

the_caller()

# Pass Function as an argument
def plus(x, y):
    return x+y

def minus(x, y):
    return x - y

def calculate(func, x, y):
    return func(x, y)

print(calculate(plus, 10, 20)) # 30
print(calculate(minus, 40, 20)) # 20


# Return function as a value
def greeting(name):
    # print("I am outer greeting.")
    def inner_greet():
        print("Hello ", name, " I am inner greeting")
    return inner_greet

new_greeting =  greeting("Subhiksha")
new_greeting()

# Python simple decorators
def make_prrety(func):
    print("Hello i am jack")
    def inner(name):
        print("Please tell me your name")
        func(name)
        print(f"Hello {name} nice to meet you.")
    return inner

def tell_the_name(name):
    print(f"Hey my name is {name}")
    return

caller = make_prrety(tell_the_name)
print(caller("Dviyanka"))

def smart_divide(func):
    def inner(a, b):
        if a == 0 or b == 0:
            print("Opps can't divide with 0")
            return
        return func(a,b)
    return inner
    
@smart_divide
def divide(a,b):
    return a/b

print(divide(12, 2))
print(divide(12, 0))
print(divide(0, 12))


def some_func(*args, **kwargs):
    print(args)
    print(kwargs)


some_func(10, [1,2,3], a=10, b=20)

def star(func):
    def inner():
        print("*" * 15)
        func()
        print("*" * 15)
    return inner


def percent(func):
    def inner():
        print("%" * 15)
        func()
        print("%" * 15)
    return inner

@percent
@star
def say_hello():
    print("HELLO")

say_hello()
```

```python
# Getter, Setter and @property decorator

class Celsius:
    def __init__(self, temperature=0):
        self.temperature = temperature

    def to_fahrenheit(self):
        return (self.temperature * 1.8) + 32
    
    # Getter and setter method for temperature.

    def get_temperature(self):
        print("Getting value..")
        return self._temperature

    def set_temperature(self, value):
        print("Setting value..")
        if value < -273.15:
            # We know that the temperature of any object cannot reach below -273.15 degrees Celsius.
            raise ValueError("Temperature below -273.15 is not possible")
        
        self._temperature = value
        return f"Value updated with {self._temperature}"
    
    temperature = property(get_temperature, set_temperature)
    

    
human = Celsius(37)
print(human.get_temperature())
print(human.to_fahrenheit())
print(human.set_temperature(-273.10))
print(human.get_temperature())
```
